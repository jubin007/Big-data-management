```python
from pyspark.streaming import StreamingContext
from pyspark.sql import Row
import time

ssc  = StreamingContext(sc, 5)
```

```python
def save(time, rdd):
    try:
        df = spark.createDataFrame(rdd.map(\
        lambda row: Row(time=time, package=row[0], count=row[1])))
        df.write.format("org.apache.spark.sql.cassandra")\
        .options(table="ggplot2_count", keyspace="streaming")\
        .save(mode="append")
    except:
        pass
```

```python
lines = ssc.textFileStream("file:///home/jubin/Downloads/streaming")
downloads_RDD = lines.map(lambda x: x.split(','))
def remove_quotation(x):
        return([xx.replace('"','')for xx in x])   
downloads_RDD = downloads_RDD.map(remove_quotation)
package_download_count = downloads_RDD.map(lambda x: (x[2], 1))
package_download_count = package_download_count.reduceByKey(lambda a,b: a + b)
package_download_count = package_download_count.filter(lambda x: "ggplot2" in x)

package_download_count.foreachRDD(save)
package_download_count.pprint()
```
